# **Data PreProcessing**

```python
#Handling missing data
data['PM2.5']=data['PM2.5'].fillna(data['PM2.5'].mean())
data['PM10']=data['PM10'].fillna(data['PM10'].mean())
data['NO']=data['NO'].fillna(data['NO'].mean())
data['NO2']=data['NO2'].fillna(data['NO2'].mean())
data['NOx']=data['NOx'].fillna(data['NOx'].mean())
data['NH3']=data['NH3'].fillna(data['NH3'].mean())
data['CO']=data['CO'].fillna(data['CO'].mean())
data['SO2']=data['SO2'].fillna(data['SO2'].mean())
data['O3']=data['O3'].fillna(data['O3'].mean())
data['Benzene']=data['Benzene'].fillna(data['Benzene'].mean())
data['Toluene']=data['Toluene'].fillna(data['Toluene'].mean())
data['Xylene']=data['Xylene'].fillna(data['Xylene'].mean())
data = data[data['AQI'].isna()==False]
```

```python
data.isnull().any()
```

```python
#Z-score normalization
import pandas as pd
import numpy as np
z_scores = np.abs((data['AQI'] - data['AQI'].mean()) / data['AQI'].std())
threshold = 2
df = data[z_scores < threshold]
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
x=df.drop(['AQI','AQI_Bucket','City','Date','State'],axis=1)
y=df['AQI']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
#Standardization
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

# **LSTM**

```python
dt = df[['CO', 'NO', 'NO2', 'O3', 'SO2', 'PM2.5', 'PM10', 'NH3','Benzene','Toluene','Xylene', 'AQI']]
dt.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
dt_normalized = scaler.fit_transform(dt)
sc2 = MinMaxScaler()
dt_y = sc2.fit_transform(df[['AQI']])
```

```python
def create_sequences(dt, seq_length):
    X, y = [], []
    for i in range(len(dt) - seq_length):
        X.append(dt[i:(i + seq_length), :-1])
        y.append(dt[i + seq_length, -1])
    return np.array(X), np.array(y)
```

```python
from sklearn.model_selection import train_test_split
train_size = int(len(dt_normalized) * 0.80)
test_size = len(dt_normalized) - train_size
train, test = dt_normalized[0:train_size, :], dt_normalized[train_size:len(dt_normalized), :]
```

```python
seq_length = 10
X_train, y_train = create_sequences(train, seq_length)
X_test, y_test = create_sequences(test, seq_length)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(units=50))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=32)
```

```python
y_pred = model.predict(X_test)
```

```python
y_test = pd.DataFrame(y_test)
y_pred = pd.DataFrame(y_pred)
pred_comp = pd.concat([y_test,y_pred],axis=1)
pred_comp.columns=['original','predicted']
pred_comp['original'] = sc2.inverse_transform(pred_comp[['original']])
pred_comp['predicted'] = sc2.inverse_transform(pred_comp[['predicted']])
pred_comp
```

```python
lstm_rmse = np.sqrt(mean_squared_error(pred_comp['original'], pred_comp['predicted']))
print("RMSE:",lstm_rmse)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

pred_comp_sample = pred_comp.head(100)

plt.plot(pred_comp_sample['original'], label='actual', color='red')
plt.plot(pred_comp_sample['predicted'], label='predicted', color='blue')
plt.legend()
plt.show()
```

# **CNN - GRU**

```python
df.head()
```

```python
df.drop(['City', 'Date', 'AQI_Bucket','State'], axis=1, inplace=True)
X = df.drop(['AQI'], axis=1)
y = df['AQI']
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_reshaped = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y, test_size=0.2, random_state=42)
```

```python
print(y_test)
```

```python
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, GRU, Dense, Flatten, RNN, SimpleRNN
from keras.optimizers import Adam
from sklearn.metrics import mean_squared_error
```

```python
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)))
model.add(MaxPooling1D(pool_size=1))
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)))
model.add(MaxPooling1D(pool_size=1))
model.add(GRU(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=16, verbose=2)
```

```python
test_preds = model.predict(X_test)
cg_rmse = np.sqrt(mean_squared_error(y_test, test_preds))
print("RMSE:", cg_rmse)
```

```python
print(test_preds)
print(test_preds.shape)
y_test_array = y_test.to_numpy()
y_test_reshaped = y_test_array.reshape(-1, 1)
print(y_test_reshaped.shape)
print(y_test_reshaped)
```
